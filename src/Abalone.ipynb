# Abalone Dataset

# Introduction
Abalone is the common name for marine snails found primarily in the UK, Austrailia, and South Africa. The age of an abalone is determined by cutting the shell through the cone, staining it, and counting the number of rings through a microscope -- a boring and time-consuming task. Other measurements, which are easier to obtain, are used to predict the age.  Here, we will try to use various abalone features to build a model for predicting abalone age.

## Data Description
These data come from a 1994 study titled: "The Population Biology of Abalone (_Haliotis_ species) in Tasmania. I. Blacklip Abalone (_H. rubra_) from the North Coast and Islands of Bass Strait".

## Continuous Attributes
- *Length*: Longest shell measurement (mm)
- *Diameter*: Perpendicular to length (mm)
- *Height*: With meat in shell (mm)
- *Whole weight*: Whole abalone (g)
- *Shucked weight*: Weight of meat (g)
- *Viscera weight*: Gut weight after bleeding (g)
- *Shell weight*: After being dried (g)

## Integer Attributes
- *Rings*: +1.5 gives the age in years

## Nominal Attributes
- *Sex*: Male (M), Female (F), and Infant (I)

## Define Custom Function

```python
######################### Skew function #########################
def skew_abalone(dataframe):
    """
    Print number of predictors at a given Skew cutoff
    """
    # Create lists for different skew levels.
    no_skew, moderate_skew, severe_skew = ([] for i in range(3))
    
    # Assign each skew score to a list.
    for i in range(len(dataframe.loc[:, dataframe.columns != 'Rings'].skew(axis=0))):
        count = dataframe.loc[:, dataframe.columns != 'Rings'].skew(axis=0)[i]
        
        # Determine skew.
        # No skew.
        if count < 0.5 and count > -0.5:
            no_skew.append(count)
        
        # Moderate skew.
        elif count > 0.5 and count < -0.5 or count < 1 and count > -1:
            moderate_skew.append(count)
        
        # Severe skew.
        else:
            severe_skew.append(count)

    # Print skew scores for each level.
    print('No Skew:', len(no_skew))
    print('Moderate Skew:', len(moderate_skew))
    print('Severe Skew:', len(severe_skew))


######################### Kurtosis function #########################
def kurtosis_abalone(dataframe):
    """
    Print number of predictors at a given Kurtosis cutoff
    """
    # Create lists for different skew levels.
    no_kurtosis, moderate_kurtosis, severe_kurtosis = ([] for i in range(3))
    
    # Assign each skew score to a list.
    for i in range(len(dataframe.loc[:, dataframe.columns != 'Rings'].kurtosis(axis=0))):
        count = dataframe.loc[:, dataframe.columns != 'Rings'].kurtosis(axis=0)[i]
        if count < 3 and count > -3:
            no_kurtosis.append(count)
        elif count > 3 and count < -3 or count < 7 and count > -7:
            moderate_kurtosis.append(count)
        else:
            severe_kurtosis.append(count)

    # Print skew scores for each level.
    print('No Kurtosis:', len(no_kurtosis))
    print('Moderate Kurtosis:', len(moderate_kurtosis))
    print('Severe Kurtosis:', len(severe_kurtosis))


######################### Regression function #########################
def regression_function(x, y):
    """
    Calculates linear model and outputs: train/validate/test shape, MSE, R2, and adjusted R2.
    """
    # Split data into training and testing sets.
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
    
    # Print size of training, validation, and testing set for verification.
    print('X_Train Shape =', round(x_train.shape[0]/4177*100), '%')
    print('X_Test Shape =', round(x_test.shape[0]/4177*100), '%')
    print('') # Add space.
    
    # Call linearRegression function from scikitlearn.
    regr = linear_model.LinearRegression()
    
    # Fit regression model using predictor(s) (X_train) and dependent variable (y_train).
    regr.fit(x_train, y_train)
    
    # Print coefficients.
    print('Coefficients')
    print(regr.coef_)
    print('') # Add space.
    
    # Make predictions using the testing set
    y_pred = regr.predict(x_test)
    
    # Print mean squared error
    print('MSE =', mean_squared_error(y_test, y_pred))
    
    # Print coefficient of determination: 1 is perfect prediction
    print('R2 =', r2_score(y_test, y_pred))
    
    # Print adjusted coefficent of determination (model steadily accrues coefficient-driven penalty, penalising complex models).
    print('Adj_R2 =', (1 - (1 - r2_score(y_test, y_pred))*((x.shape[0] - 1)/(x.shape[0] - x.shape[1] - 1))))
    
    return regr, x_train, y_train


######################### Function for testing linearity #########################
def calculate_residuals(model, features, label):
    """
    Creates predictions on the features with the model and calculates residuals
    """
    # vector of predicted response variables.
    predictions = model.predict(features)
    
    # Dataframe plotting labels against model predictions.
    df_results = pd.DataFrame({'Actual': label, 'Predicted': predictions})
    
    # Absolute difference between labels and predictions.
    df_results['Residuals'] = abs(df_results['Actual']) - abs(df_results['Predicted'])
    
    return df_results


######################### Funcion for testing normality #########################
def normal_errors_assumption(model, features, label, p_value_thresh = 0.05):
    """
    Normality: Assumes that the error terms are normally distributed. If they are not,
    nonlinear transformations of variables may solve this.
               
    This assumption being violated primarily causes issues with the confidence intervals
    """
    from statsmodels.stats.diagnostic import normal_ad
    print('Assumption 2: The error terms are normally distributed', '\n')
    
    # Calculating residuals for the Anderson-Darling test
    df_results = calculate_residuals(model, features, label)
    
    print('Using the Anderson-Darling test for normal distribution')

    # Performing the test on the residuals
    p_value = normal_ad(df_results['Residuals'])[1]
    print('p-value from the test - below 0.05 generally means non-normal:', p_value)
    
    # Reporting the normality of the residuals
    if p_value < p_value_thresh:
        print('Residuals are not normally distributed')
    else:
        print('Residuals are normally distributed')
    
    # Plotting the residuals distribution
    plt.subplots(figsize=(12, 6))
    plt.title('Distribution of Residuals')
    sns.distplot(df_results['Residuals'])
    plt.show()
    
    print()
    if p_value > p_value_thresh:
        print('Assumption satisfied')
    else:
        print('Assumption not satisfied')
        print()
        print('Confidence intervals will likely be affected')
        print('Try performing nonlinear transformations on variables')


######################### Polynomial regression function #########################
def PolynomialRegression(degree = 2, **kwargs):
    """
    Create a pipeline for polynomail regression.
    """
    # Return polynomial pipeline.
    return make_pipeline(PolynomialFeatures(degree), LinearRegression(**kwargs))


######################### Split Rings function #########################
def abalone_split(dataframe, split):
    """
    Splits dependent variable in 2 or 3 groups.
    """
    # Copy abalone dataframe.
    abalone_two = abalone.copy()
    
    # Empty list.
    Rings = []
    
    # Split by 2.
    if split == 2:
        # Perform mean split: Rings list stores 1 for abalone with more than 14.5 rings, and 0 otherwise.
        for i in abalone_two['Rings']:
            if i > 14.5:
                Rings.append('1')
            else:
                Rings.append('0')
   
    # Split by 3.            
    else:
        for i in abalone_two['Rings']:
            if i > 19.2:
                Rings.append('2')
            elif i > 9.6 and i < 19.2:
                Rings.append('1') 
            else:
                Rings.append('0')
                
    # Replace total Rings column with mean-split Rings column.
    abalone_two['Rings'] = Rings

    # See how many of each their are.
    abalone_two['Rings'].value_counts()
    
    # Return dataframe if split is 2 or 3.
    if split == 2 or split == 3:
        return(abalone_two)
    
    # Else, tell user to use split of 2 or 3.
    else:
        print("Can only split by 2 or 3. Please try again!")


######################### Scaled Regression function #########################
def scale_regression_function(X, y):
    """
    Calculates linear model and outputs: train/validate/test shape, MSE, R2, and adjusted R2.
    """
    
    num_pipeline = Pipeline([
    ('std_scaler', StandardScaler())
    ])
    
    X_scaled = num_pipeline.fit_transform(X)
    
    # Split data into training and testing sets.
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.2, random_state = 1)

    # further divide the training data again to get a validation set.
    X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 0.25, random_state = 1) # 0.25 x 0.8 = 0.2
    
    # Print size of training, validation, and testing set for verification.
    print('X_Train Shape =', round(X_train.shape[0]/4177 * 100),'%')
    print('X_Validate Shape =', round(X_val.shape[0]/4177 * 100),'%')
    print('X_Test Shape =', round(X_test.shape[0]/4177 * 100),'%')
    print('') # Add space.
    
    # Call linearRegression function from scikitlearn.
    regr = linear_model.LinearRegression()
    
    # Fit regression model using predictor(s) (X_train) and dependent variable (y_train).
    regr.fit(X_train, y_train)
    
    # Print coefficients.
    print('Coefficients')
    print(regr.coef_)
    print('') # Add space.
    
    # Make predictions using the testing set
    y_pred = regr.predict(X_val)
    
    # Print mean squared error
    print('MSE =', mean_squared_error(y_val, y_pred))
    
    # Print coefficient of determination: 1 is perfect prediction
    print('R2 =', r2_score(y_val, y_pred))
    
    # Print adjusted coefficent of determination (model steadily accrues coefficient-driven penalty, penalising complex models).
    print('Adj_R2 =', (1 - (1 - r2_score(y_val, y_pred)) * ((X.shape[0] - 1) / (X.shape[0] - X.shape[1] - 1))))
```

We cannot use objects in a regression model, as python will give us an error. But we can't just replace the 3 unique values with unique digits. When you do this, the model assumes these refer to ranks. Here, Sex is not ranked, it is just an indicator. Thus,  we need to make 3 columns, each with a 1 to indicate that row is part of that Sex, and a 0 otherwise.

First, let's get a list of all categorical variables to make sure they are only contained in the Sex column.

## Linearity
There should be a linear association between the model predictors, and the response/dependent variable. Linear regression (and modelling in general) is about explaining a process, and if the dependent variable is not alinear combination of predictors, then the linear regression would be a poor model. This can produce very innacurate predictions because the model will be underfitted. We can check this by plotting all model's predicted values against the actual values (i.e. see the residuals). A perfectly linear association would thus be indicated by a perfectly straight line: Every predicted value would be the actual value.

```python
# Calculate residuals.
df_results = calculate_residuals(regr, X_train, y_train)
    
# Plot the actual vs predicted values.
sns.lmplot(x='Actual', y='Predicted', data=df_results, fit_reg=False, size=7)
        
# Plot a diagonal line.
line_coords = np.arange(df_results.min().min(), df_results.max().max())
plt.plot(line_coords, line_coords,  # X and y points
             color='darkorange', linestyle='--')
plt.title('Actual vs. Predicted')
plt.show()
```

Is the relationship linear? It's not perfect, but not to bad. As the actual values get large, the model seems to underpredict them.

## Learning Curve

```python
# Select sample sizes of training set
train_sizes = [1, 251, 501, 752, 1002, 1253, 1503, 1754, 2004, 2255, 2505]

features = ['length', 'height', 'shucked_weight', 'shell_weight']
target = 'rings'
train_sizes, train_scores, validation_scores = learning_curve(
    estimator = LinearRegression(), 
    X = abalone[features],
    y = abalone[target],
    train_sizes = train_sizes,
    cv = 5,
    scoring = 'neg_mean_squared_error')

train_scores_mean = -train_scores.mean(axis = 1)
validation_scores_mean = -validation_scores.mean(axis = 1)
print('Mean training scores\n\n', pd.Series(train_scores_mean, index = train_sizes))
print('\n', '-' * 20) # separator
print('\nMean validation scores\n\n',pd.Series(validation_scores_mean, index = train_sizes))
```
